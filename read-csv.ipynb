```python
import pandas as pd

# Read both CSV files
long_df = pd.read_csv('1_film-dataset_festival-program_long.csv')
wide_df = pd.read_csv('1_film-dataset_festival-program_wide.csv')

print(f"Long file: {len(long_df)} rows")
print(f"Wide file: {len(wide_df)} rows")
```

```python
# Merge on unique.id - this adds all columns from wide_df to long_df
merged_df = long_df.merge(wide_df, on='unique.id', how='left')

print(f"Merged file: {len(merged_df)} rows")
print(f"\nColumns: {list(merged_df.columns)}")
```

```python
# Preview the merged data (showing key columns)
merged_df[['film.id', 'unique.id', 'imdb.id', 'title.mixed', 'fest', 'fest.sect']].head(10)
```

```python
# Save to a new CSV file
merged_df.to_csv('merged_film_dataset.csv', index=False)
print('Saved to merged_film_dataset.csv')
```
